```python
import subprocess
import time

print("Starting Streamlit app with localtunnel...")

# Start localtunnel and capture its output for the URL
st_tunnel = subprocess.Popen('npx localtunnel --port 8501', shell=True, stdout=subprocess.PIPE)
time.sleep(4) # Give localtunnel some time to establish the connection and print the URL
streamlit_link = st_tunnel.stdout.readline().decode('utf-8').strip()
print(f"Streamlit Link: {streamlit_link}")

# Start the Streamlit app in the background
subprocess.Popen("streamlit run app/Home.py --server.port 8501", shell=True)

print("\n✅ Streamlit app and localtunnel should be running.")
print("Access your Streamlit app at the link above.")
```

```python
import subprocess
import time
import os

# 1. Verify the DB exists before starting
db_path = "/content/drive/MyDrive/PatrolQ/mlflow.db"
if not os.path.exists(db_path):
    print(f"⚠️ Warning: Could not find DB at {db_path}")
    print("Check if Drive is mounted and the path is correct.")

# 2. Get Bypass IP
print("Bypass IP:")
!curl -s ipv4.icanhazip.com

# 3. Start Tunnels
print("\nStarting Tunnels...")
mlflow_tunnel = subprocess.Popen('npx localtunnel --port 5000', shell=True, stdout=subprocess.PIPE)
time.sleep(4)
print(f"MLflow Link: {mlflow_tunnel.stdout.readline().decode('utf-8').strip()}")

st_tunnel = subprocess.Popen('npx localtunnel --port 8501', shell=True, stdout=subprocess.PIPE)
time.sleep(4)
print(f"Streamlit Link: {st_tunnel.stdout.readline().decode('utf-8').strip()}")

# 4. Start Servers
# We use sqlite:/// to tell MLflow to use your specific database file
subprocess.Popen(f"mlflow ui --backend-store-uri sqlite:///{db_path} --port 5000", shell=True)

# Start Streamlit and capture stderr for debugging
streamlit_process = subprocess.Popen(
    "streamlit run app/Home.py --server.port 8501",
    shell=True,
    stderr=subprocess.PIPE,
    stdout=subprocess.PIPE
)

# Give Streamlit a moment to start and produce output
time.sleep(5)

# Read any error output from Streamlit
stderr_output = streamlit_process.stderr.read().decode('utf-8')
if stderr_output:
    print("\nStreamlit Error Output:")
    print(stderr_output)

print("\n✅ Systems Online. Give it 15 seconds to load your DB data.")
```
